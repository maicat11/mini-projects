### Problem 1
Given a list of timestamps in sequential order, return a list of lists grouped by week (7 days) using the first timestamp as the starting point.

Example:   

ts = [
    '2019-01-01', 
    '2019-01-02',
    '2019-01-08', 
    '2019-02-01', 
    '2019-02-05',
]

output = [
    ['2019-01-01', '2019-01-02'], 
    ['2019-01-08'], 
    ['2019-02-01', '2019-02-05'],
]

```python
import pandas as pd
from datetime import date, timedelta

dates = pd.read_csv('data/dates.txt')

dates['date'] = pd.to_datetime(dates['date'], format='%Y-%m-%d')
dates['date'] = dates['date'].dt.date
print(dates.shape)
dates.head()
```

```python
week_end = dates['date'][0] + timedelta(days=7)
all_weeks = []
weekly = []

for d in dates['date']:
    if d < week_end:
        weekly.append(d.strftime('%Y-%m-%d'))
    elif d > week_end:
        week_end += timedelta(days=7)
        all_weeks.append(weekly)
        weekly = []
all_weeks
```
